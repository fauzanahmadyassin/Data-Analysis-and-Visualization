### Data Analysis and Visualization on iPhone Product from Mudah.my

All data has been scraped from https://www.mudah.my/Malaysia/Mobile-Phones-and-Gadgets-3020/iphone-all. <br>
Data scraping process was set to scrape only from 200 out of 512 pages <br> 

#### Import related library

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

list_name = []
list_price = []
list_area = []
list_cond = []
list_seller = []
```

#### Data scraping process

```python
i=1
url = "https://www.mudah.my/Malaysia/Mobile-Phones-and-Gadgets-3020/iphone-all?o="+str(i)+"&q=iphone&th=1"

while (i<200):

    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'lxml')

    phone_name = soup.find_all('h2',{'class':'list_title'})
    phone_price = soup.find_all('div',{'class':'ads_price'})
    area = soup.find_all('div',{'class':'area'})
    condition = soup.find_all('font',{'class':'icon_label'})
    seller = soup.find_all('div',{'class':'container_ico'})
    
    
    phone_name = [a.text for a in phone_name]
    phone_name = phone_name[:-1]
    
    phone_price = [a.text for a in phone_price]
    phone_price = phone_price[:-1]
    
    area = [a.text for a in area]
    area = area[:-1]
    
    condition = [a.text for a in condition]
    condition = condition[:-1]
    
    seller = [a.text for a in seller]
    seller = seller[:-1]
   

    list_name = list_name + phone_name
    list_price = list_price + phone_price
    list_area = list_area + area
    list_cond = list_cond + condition
    list_seller = list_seller + seller
    
    print ('Data from page '+str(i)+' has been scraped')
    
    i = i+1    
    url = "https://www.mudah.my/Malaysia/Mobile-Phones-and-Gadgets-3020/iphone-all?o="+str(i)+"&q=iphone&th=1"
    #print (url)
    
 
```

<br> 
#### Checking the data volume from all columns
<br>As we can see the number of sample is different from each columns 

```python
df_name = pd.DataFrame()
df_price = pd.DataFrame()
df_area = pd.DataFrame()
df_cond = pd.DataFrame()
df_seller = pd.DataFrame()

print (len(list_name))
print (len(list_price))
print (len(list_area))
print (len(list_cond))
print (len(list_seller))

df_name['phone_name'] = list_name
df_price['phone_price'] = list_price
df_area['area'] = list_area
df_cond['condition'] = list_cond
df_seller['seller'] = list_seller

```

#### Displaying the data into table form


```python
df_name = pd.concat([df_name, df_price, df_area, df_cond, df_seller], axis=1)

df_name
```

#### Remove all the rows that contain null or NaN values

```python
df_name = df_name.dropna()

df_name
```

#### Remove white space before and after each string

For example replacing  <i>'&nbsp;&nbsp;&nbsp;&nbsp;Private seller&nbsp;&nbsp;'</i> to <i>'Private seller'</i>

```python
df = pd.DataFrame()

df[df_name.columns] = df_name.apply(lambda x: x.str.strip())

df
```

```python
df_x = pd.DataFrame()
df_xs = pd.DataFrame()
df_5 = pd.DataFrame()
df_5s = pd.DataFrame()
df_6 = pd.DataFrame()
df_6s = pd.DataFrame()
df_6splus = pd.DataFrame()
df_7 = pd.DataFrame()
df_7plus = pd.DataFrame()
df_se = pd.DataFrame()


df_x = df[df['phone_name'].str.contains('iphone x ', case=False)]
df_xs = df[df['phone_name'].str.contains('iphone xs', case=False)]
df_5 = df[df['phone_name'].str.contains('iphone 5 ', case=False)]
df_5s = df[df['phone_name'].str.contains('iphone 5s', case=False)]
df_6 = df[df['phone_name'].str.contains('iphone 6 ', case=False)]
df_6s = df[df['phone_name'].str.contains('iphone 6s', case=False)]
df_6splus = df[df['phone_name'].str.contains('iphone 6s plus', case=False)]
df_7 = df[df['phone_name'].str.contains('iphone 7 ', case=False)]
df_7plus = df[df['phone_name'].str.contains('iphone 7 plus', case=False)]
df_se = df[df['phone_name'].str.contains('iphone se', case=False)]

```

#### Displaying the graph of total number of iPhones 

```python
import matplotlib.pyplot as plt

phone = ('iPhone Xs', 'iPhone X','iPhone 7 plus','iPhone 7','iPhone 6s Plus', 'iPhone 6s', 'iPhone 6', 'iPhone SE','iPhone 5s','iPhone 5')
y_pos = np.arange(len(phone))
performance = [len(df_xs), len(df_x), len(df_7plus), len(df_7), len(df_6splus), len(df_6s), len(df_6), len(df_se), len(df_5s), len(df_5)]
 
    
plt.figure(dpi=150, figsize=(12,6))
plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, phone)
plt.ylabel('Count\n')
plt.xlabel('\nPhone model')
plt.title('\n\nTotal Number of iPhones for being sold in Mudah.my\n ')
 
plt.show()
```

#### Seller and condition graph visualization in percentage

```python
labels1 = df_seller.seller[0], df_seller.seller[1]
sizes1 = df_seller.phone_name[0], df_seller.phone_name[1]

labels2 = df_cond.condition[0], df_cond.condition[1], df_cond.condition[2]
sizes2 =  df_cond.phone_name[0],  df_cond.phone_name[1], df_cond.phone_name[2]

fig, (ax1, ax2) = plt.subplots(1, 2, dpi=120, figsize=(5.5,2.5), sharex=True)

ax1.pie(sizes1, labels=labels1, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax2.pie(sizes2, labels=labels2, autopct='%1.1f%%',
        shadow=False, startangle=90)

plt.show()
```
